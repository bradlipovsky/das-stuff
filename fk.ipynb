# Frequency-Wavenumber analysis of DAS data

```python
import matplotlib.pyplot as plt
import numpy as np
from dasquakes import *
```

## Example 1: Scholte Waves in the Puget Sound

```python
ft,f,k = fk_analysis(t0=datetime.datetime(2022, 5, 17, 11, 16, 0),draw_figure=False,
                     downsamplefactor=1,
                     record_length=2)
```

```python
plt.subplots(figsize=(14,10))
c=plt.pcolormesh(k,f,np.log10(np.abs(ft)),vmin=1,vmax = 6)
plt.colorbar(c)
plt.xlabel('Wavenumber (1/m)')
plt.ylabel('Frequency (Hz)')
plt.ylim([-4,4])
plt.show()
```

## Example 2: An earthquake recorded in N. Seattle

```python
ft,f,k = fk_analysis(t0=datetime.datetime(2022, 5, 1, 11, 16, 0),draw_figure=False,
                     cable='seadasn', #2022-05-01_11-15
                     downsamplefactor=1,
                     record_length=2,
                     channel_range=[350,600])
```

```python
plt.subplots(figsize=(14,10))
c=plt.pcolormesh(k,f,np.log10(np.abs(ft)),vmin=0,vmax = 4)
plt.plot([0,0.04],[0,15])
plt.colorbar(c)
plt.xlabel('Wavenumber (1/m)')
plt.ylabel('Frequency (Hz)')
plt.ylim([-20,20])
plt.show()
```
