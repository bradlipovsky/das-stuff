## What does the Sintela Onyx measure

The answer is: Spatial Differential of Optical Phase \
While most DAS systems do this, the output of different DAS systems is in different units: E.g. the Silixa iDAS outputs strain-rate.\
How does the spatial differential of optical phase relate to strain (that's what we want to measure here)?\
Well, that's actually great, because it is directly proportional to strain/displacement. (Actually I think it is the proportional to the strain per gauge length per time... So it's not so much different to the Silixa iDAS.)

From Sintela:\
$ \epsilon = \phi \cdot \frac{\lambda}{4 \pi \ \varepsilon \ n_{eff} l_{gauge}}$

$\epsilon$: strain\
$\phi$: spatial differential of optical phase (what we measure)\
$\lambda$: vacuum wavelength of laser in Onyx (1550.12nm)\
$\varepsilon$: strain optic coefficient (0.78)\
$n_{eff}$: effective refractive index of fiber (1.4682) !!!check this!!!\
$l_{gauge}$ gauge length

```python
# define function for converting to strain

# define global variables
laser_wavelength = 1550.12*1e-9 #[m]
opt_coeff = 0.78
ref_idx = 1.4682
l_gauge = attrs['GaugeLength']

def phi2strain(phi_data):
    '''input array type object with phase measurements'''
    strain = phi_data * laser_wavelength / (4 * np.pi * opt_coeff * ref_idx * l_gauge)
    return(strain)
```

```python
# convert raw data to strain (per time per gauge length)
s_filt = phi2strain(data_filt)
s = phi2strain(data_select)
```

```python
# look into the mean/median strain over time per cannel
plt.plot(np.mean(s_filt, axis=0))
plt.plot(np.median(s_filt, axis=0), linestyle='dashed')
plt.ylabel('Strain per time per gauge length')
plt.xlabel('Selected Channels (not real channel number)')
```

See if we get any numbers of strain over time and space that make sense

```python
# how does the cummulative strain develop over time per each channel
t_secs = np.arange(0,nt/attrs['PulseRate'],1/attrs['PulseRate'])
for i in range(np.shape(s_filt)[1]):
    plt.plot(t_secs, np.cumsum(s_filt[:,i], axis=0))
plt.xlabel('Seconds')
plt.ylabel('Strain per Channel')
```

Hmmmm, I would say something in the range of millimeters over the length of 10m would be in the range of how much the cable could be strained, but it's a lot. But ok, we expect that not all channels are performing super good! Maybe it is worth getting rid of them. That would essentially mean that we could e.g. only look at some median value. Maybe even the mean is not too bad.

```python
# let's look at all channels together
s_chamean = np.cumsum(np.mean(s_filt, axis=1))
s_chamedian = np.cumsum(np.median(s_filt, axis=1))

t_secs = np.arange(0,s_filt.shape[0]/attrs['PulseRate'],1/attrs['PulseRate'])
plt.plot(t_secs, s_chamean, label='mean', color='C0')
plt.plot(t_secs, s_chamedian, label='median', color='C1')
# plot RMS of drift over time
s_rms_mean = np.sqrt(np.mean(s_chamean**2))
plt.hlines(np.array([-s_rms_mean, s_rms_mean]), 0,t_secs[-1], color='C0', linestyle='dashed')
s_rms_median = np.sqrt(np.mean(s_chamedian**2))
plt.hlines(np.array([-s_rms_median, s_rms_median]), 0,t_secs[-1], color='C1', linestyle='dashed')

plt.xlabel('Time [s]')
plt.ylabel('Strain')
plt.legend()
```

Carefull here! The median is not really something physical here, because it only picks the value of any of the selected channels, that lies in the center of the strain distribution at a given time. But maybe it tells us something! Overall strains of $10^{-5}$ seems high, but still reasonable over one minute.

Now do this for a longer time. But wait, we need to make sure, that we are not suffering from filter artefacts at the boundaries! So let's look, how the strain evolution looks for unfiltered data.

```python
s_chamean = np.cumsum(np.mean(s, axis=1))
s_chamedian = np.cumsum(np.median(s, axis=1))

t_secs = np.arange(0,s.shape[0]/attrs['PulseRate'],1/attrs['PulseRate'])
plt.plot(t_secs, s_chamean, label='mean', color='C0')
plt.plot(t_secs, s_chamedian, label='median', color='C1')
# plot RMS of drift over time
s_rms_mean = np.sqrt(np.mean(s_chamean**2))
plt.hlines(np.array([-s_rms_mean, s_rms_mean]), 0,t_secs[-1], color='C0', linestyle='dashed')
s_rms_median = np.sqrt(np.mean(s_chamedian**2))
plt.hlines(np.array([-s_rms_median, s_rms_median]), 0,t_secs[-1], color='C1', linestyle='dashed')

plt.xlabel('Time [s]')
plt.ylabel('Strain')
plt.legend()
```

Great! That looks similar to the plot above. I think for now, we should continue with unfiltered data. This makes things much easier, since we can simply read the data in file by file and calculate the cummulative strain.

What is the goal here? \
Well in the best case, we see some variation of strain with temperature (as we see it for broadband seismometers). Daily temperature variations are a neat measure to evaluate the response of DAS on strain.

save only difference from 0 to end of minute. That should be sufficient to have a smooth day plot (plot 2 days)

```python
# loop over two days of data

# select cable and record length to read 
cable='seadasn'
record_length = 1 #minutes

t_start = datetime.datetime(2022, 11, 4, 0, 0, 0)
t_end = datetime.datetime(2022, 11, 4, 4, 0, 0)

# select channels
start_cha = 350
end_cha = 800

# duration in minutes
t_diff = t_end-t_start
minutes = int(t_diff.total_seconds() / 60)

diff_arr = np.full(minutes,np.nan)

for i in range(minutes):
    
    # get data storage information
    t = t_start+datetime.timedelta(minutes=i)
    prefix, network_name, datastore = data_wrangler(cable,record_length,t0)
    
    # read data
    data,dates,attrs = open_sintela_file(prefix,
                                     t,
                                     datastore,
                                     number_of_files=1,
                                     verbose=False)
    
    # select channels to work with
    data_select = data[:,start_cha:end_cha]
    
    # convert to strain
    strain = phi2strain(data_select)
    
    # sum over channel mean strain over all times
    # Attention! here we neglegt the offset between two files 
    #(this solution here is not accurate, but much easier)
    s_min_sum = np.sum(np.mean(strain, axis=1))
    diff_arr[i] = s_min_sum
    
    sys.stdout.write('\rMinute: {}'.format(i))
    sys.stdout.flush()
```

```python
# Plot evolution of strain in cable over time
plt.plot(np.cumsum(diff_arr))
plt.xlabel('Time [s]')
plt.ylabel('Strain')
```

This is highly parallelizable!

```python
import itertools
import sys
import multiprocessing

# loop over two days of data

# select cable and record length to read 
cable='seadasn'
record_length = 10 #minutes

t_start = datetime.datetime(2022, 11, 4, 0, 0, 0)
t_end = datetime.datetime(2022, 11, 4, 4, 0, 0)

# select channels
start_cha = 350
end_cha = 800

# duration in minutes
t_diff = t_end-t_start
minutes = int(t_diff.total_seconds()/60)
minutes_list = np.arange(0,minutes,record_length)

diff_arr = np.full(minutes,np.nan)
diff_arr_cum = np.full((len(minutes_list),int(record_length*60*attrs['PulseRate'])),np.nan)

def calc_strain(minute):
    # get data storage information
    t = t_start+datetime.timedelta(minutes=int(minute))
    prefix, network_name, datastore = data_wrangler(cable,record_length,t0)
    
    # read data
    data,dates,attrs = open_sintela_file(prefix,
                                     t,
                                     datastore,
                                     number_of_files=record_length,
                                     verbose=False)
    
    # select channels to work with
    data_select = data[:,start_cha:end_cha]
    
    # convert to strain
    strain = phi2strain(data_select)
    
    # sum over channel mean strain over all times
    # Attention! here we neglegt the offset between two files 
    #(this solution here is not accurate, but much easier)
    s_min_sum = np.sum(np.mean(strain, axis=1))
    s_min_cumsum = np.mean(strain, axis=1)
    return((s_min_sum,s_min_cumsum))

# main program
import time
s_time = time.time()
p = multiprocessing.Pool(processes=50)
# parallel reading and analizing of data
for i, s_return in enumerate(p.imap(calc_strain, 
                              minutes_list),0):
    diff_arr[i] = s_return[0]
    diff_arr_cum[i] = s_return[1]
    sys.stdout.write('\r{} of {}'.format(np.count_nonzero(~np.isnan(diff_arr)), len(minutes_list)))
    sys.stdout.flush()
p.close()
p.join()
print('\relapsed time: {:.0f} s'.format(time.time()-s_time))
```

```python
# Plot evolution of strain in cable over time
plt.plot(np.cumsum(diff_arr_cum.flatten()))
plt.xlabel('Time [s]')
plt.ylabel('Strain')
```
