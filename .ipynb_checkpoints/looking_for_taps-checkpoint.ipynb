```python
file = '/data/data7/seadasn/seadasn_2022-10-05_01-14-00_UTC_038110.h5'
```

```python
import h5py
```

```python
data = h5py.File(file,'r')
```

```python
import numpy as np
phi = np.array(data['Acquisition/Raw[0]/RawData'])
```

```python
phi.shape
```

```python
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
[b,a]=butter(2,10/50,'high')
phi_filt = filtfilt(b,a,phi,axis=0)
```

```python
plt.subplots(figsize=(10,5))
vm = 0.01
c=plt.imshow(phi_filt,extent=[0,29.9,0,60],vmin=-vm,vmax=vm,aspect='auto')
plt.colorbar(c)
plt.xlabel('Optical Distance (km)')
plt.ylabel('Time (s)')
plt.xlim([4,5])
plt.show()
```

```python
plt.subplots()
plt.plot(phi_filt[:,730:731])
plt.xlim([0,1200])
plt.show()
```
