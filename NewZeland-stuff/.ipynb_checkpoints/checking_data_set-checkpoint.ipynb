```python
import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
import glob
```

```python

file_list = glob.glob("/data/data1/Ruapehu/decimator_ruapehu/decimator1_2023-04-02_08.45*")
file_list
```

```python
#event of interest 
#Public ID: 2023p246820
#Time: 2023-04-02 19:46:01 (NZST)
#Magnitude: 1.1
#Depth: 4 km
data = np.array([])
#time = np.array([])

data1 = h5py.File('/data/data1/Ruapehu/decimator_ruapehu/decimator1_2023-04-02_08.45.19_BST.h5','r')
data2 = h5py.File('/data/data1/Ruapehu/decimator_ruapehu/decimator1_2023-04-02_08.46.19_BST.h5','r')
```

```python
chan_min=0
chan_max=-1
```

```python
this_data = np.array(data1['Acquisition/Raw[0]/RawData'][:,chan_min:chan_max])
this_time = np.array(data1['Acquisition/Raw[0]/RawDataTime'])

this_data2 = np.array(data2['Acquisition/Raw[0]/RawData'][:,chan_min:chan_max])
this_time2 = np.array(data2['Acquisition/Raw[0]/RawDataTime'])

data = np.concatenate((this_data, this_data2 ))
time = np.concatenate((this_time, this_time2 ))
```

```python
attrs=dict(data1['Acquisition'].attrs)
attrs
```

```python

import matplotlib.dates as mdates

#date_format = mdates.DateFormatter('%H:%M:%S')

date_format = mdates.DateFormatter('%H:%M:%S')
x_lims1 = mdates.date2num(time)
x_max = data.shape[1] * attrs['SpatialSamplingInterval']
fig,ax=plt.subplots(figsize=(15,8))
plt.imshow(this_data.T,cmap='seismic',aspect='auto',vmin=-0.1,vmax=0.1,extent=[x_lims1[0],x_lims1[-1],x_max,0])
plt.title('Raw data-Das event',fontsize=15)
plt.xlabel("Time (s)-UTC",fontsize=20)
plt.ylabel("Distance along the cable (m) ", fontsize=20)

ax.xaxis.set_major_formatter(date_format)
ax.xaxis_date()
plt.grid()
plt.show()
```

```python
low_cut1 = 1.5
hi_cut1 = 8

b,a = butter(2,(low_cut1,hi_cut1),'bp',fs=attrs['MaximumFrequency']*2)
data_filt1 = filtfilt(b,a,data,axis=0)
```

```python
date_format = mdates.DateFormatter('%H:%M:%S')
x_lims1 = mdates.date2num(time)
x_max = data_filt1.shape[1] * attrs['SpatialSamplingInterval']
fig,ax=plt.subplots(figsize=(10,5))
plt.imshow(data_filt1.T,cmap='seismic',aspect='auto',vmin=-0.1,vmax=0.1,extent=[x_lims1[0],x_lims1[-1],x_max,0])
plt.title('Filter data-DAS event from NZ', fontsize=15)
plt.xlabel("Time-UTC", fontsize=20)
plt.ylabel("distance along the cable (m) ", fontsize=20)

ax.xaxis.set_major_formatter(date_format)
ax.xaxis_date()
plt.grid()
plt.show()
```

```python
plt.plot(x_lims1,data_filt1[:,100],label='channel 100')
plt.plot(x_lims1,data_filt1[:,300],label='channel 300')
plt.xlabel('time')
plt.ylabel('phase (radians)')
plt.show()
```

```python
from obspy.core import UTCDateTime
from obspy.clients.fdsn import Client

arc_client  = 'http://service.geonet.org.nz'
# or arc_client = "GEONET"
nrt_client = 'http://service-nrt.geonet.org.nz'
```

```python
t = UTCDateTime('2023-04-02')

#first try to get data from archive server
try:
    client = Client(arc_client)
    st = client.get_waveforms('NZ', 'TRVZ', '10', 'HHZ', t+27950, t + 28000,attach_response=True)
    print('arc client successful')
#if this raises an exception, try the near real-time server
except:
    client = Client(nrt_client)
    st = client.get_waveforms('NZ', 'TRVZ', '10', 'HHZ', t+27950, t + 28000,attach_response=True)
    print('nrt client successful')
print(st)
```

```python
low_cut1 = 1.5
hi_cut1 = 8
pre_filt = (low_cut1/2, low_cut1, hi_cut1, hi_cut1*2)
st.remove_response(output='VEL', pre_filt=pre_filt)
st.plot() 

```
